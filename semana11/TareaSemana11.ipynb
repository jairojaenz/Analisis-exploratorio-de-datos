```python
# Importar librer√≠as
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# En base al archivo tips.csv
# Leer el archivo
tips = sns.load_dataset('tips')
# Mostrar las primeras filas
print(tips.head())
```

```python
# scatterplot
sns.relplot(
    kind='scatter',
    x='total_bill',
    y='tip',
    data=tips
)
plt.show()
print('Grafico sin segmentacion categorica')
```

```python
# scatterplot con segmentaci√≥n categ√≥rica
sns.relplot(
    kind='scatter',
    x='total_bill',
    y='tip',
    data=tips,
    hue='time'
)
plt.show()
print('Grafico con segmentacion categorica')


```

```python
# joinplot
sns.jointplot(
    x='total_bill',
    y='tip',
    data=tips,
    hue='time'
)
```

```python
# Boxplot
sns.boxplot(
    x='time',
    y='tip',
    data=tips
)
plt.show()
```

```python
# barplot
sns.barplot(
    x='time',
    y='tip',
    data=tips
)
plt.show()
```

```python
# Pairplot
sns.pairplot(
    data=tips,
    corner=True,
    hue='sex',
    kind='scatter'
)
plt.show()
```
