```python
# usaremos pandas y seaborn para visualizar y lecutra de dataframes
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset iris.csv
iris = sns.load_dataset('iris')
iris
```

```python
# scatter plot
sns.relplot(
    kind='scatter',
    data=iris,
    x='sepal_length',
    y='petal_length',
)
print('Grafico sin segmentacion categorica')
plt.show()
```

```python
sns.relplot(
    kind='scatter',
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species',
)
print('Grafico con segmentacion categorica')
plt.show()
print('Al principio no se veia la diferencia entre las especies, pero al segmentar por especie se puede ver que las especies se distribuyen de manera diferente en el grafico')
```

```python
# Programando un joinplot
sns.jointplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species',
)
print('Trabajando con jointplot')
```

```python
# Programanndo un boxplot
sns.boxplot(
    data=iris,
    x='species',
    y='sepal_width',
)
```

```python
# Programanndo un barplot
sns.barplot(
    data=iris,
    x='species',
    y='sepal_width',
)
```

```python
# Programanndo un pairplot
sns.pairplot(
    data=iris,
    hue='species',
    corner=True,
    kind='scatter',
)
plt.show()
print('Podemos ayudarnos de pairplot para ver la relacion entre las variables de nuestro dataset, en este caso se segmenta por especie para ver como se distribuyen las variables en el grafico')
```

```python
# Matriz de covarianza
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
iris = sns.load_dataset('iris')
iris
```

```python
# Visualizacion rapida de las correlaciones
sns.pairplot(iris, hue='species', corner=True, kind='scatter')
```

```python
iris.columns
```

```python
# Normalizar los datos
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
iris_scaled
```

```python
# paso 2 hallar la matriz trasnpuesta
iris_scaled.T
```

```python
#  paso 3 calcular las covarianzas de las variables
cov_matrix = np.cov(iris_scaled.T) # T es la matriz trasnpuesta
cov_matrix # matriz de covarianza
```

```python
# paso 4 graficar la matriz de covarianza
plt.figure(figsize=(10, 10)) # tamaÃ±o de la figura
sns.set(font_scale=1.5) # tamaÃ±o de la fuente
sns.heatmap(
                data=cov_matrix, # matriz de covarianza
                 cbar=True, # barra de color
                 annot=True, # anotacion de los valores
                 square=True, # cuadrado
                 fmt='.5f', # formato de los valores
                 annot_kws={'size': 15}, # tamaÃ±o de la anotacion
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], # etiquetas del eje y
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']) # etiquetas del eje x
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Cargar el dataset iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# verificar los nombres de las columnas
iris_df.columns

# seleccionar las columnas numÃ©ricas usando los nombres correctos
x = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
# definir la matriz de covarianza manualmente
cov_matrix = x.cov()
# Definir metodos para calcular la matriz de covarianza/ correlacion
methods = {
    'cov_matrix':cov_matrix, # matriz de covarianza
    'Iris.corr()': x.corr(), # matriz de correlacion
    'Ecuacion manual de la correlacion': ((x-x.mean())/x.std()).cov().T, # matriz de covarianza manual
}
# crear subplots
fig, axs = plt.subplots(1, 3, figsize=(40, 12), sharey=True, sharex=True)
fig.suptitle('Diferencias en distintos metodos de calcular la matriz de covarianza')
# graficar cada matriz de covarianza
cont = 0
for title, method in methods.items():
    axs[cont].set_title(title)
    sns.heatmap(
        data=method,
        cbar=True,
        annot=True,
        square=True,
        fmt='.2f',
        annot_kws={'size': 10},
        yticklabels=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
        xticklabels=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
        ax=axs[cont]
    )
    cont += 1
plt.show()

    
```
