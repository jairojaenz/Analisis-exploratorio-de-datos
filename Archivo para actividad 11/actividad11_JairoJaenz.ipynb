```python
# En esta actividad deberán realizar un bivariado y multivariado para los 
# datasets “exercise.csv” y “car_crashes.csv” para lo cual harán uso de: 
# 1.	Scatterplot (diagrama de dispersión)
# 2.	Joinplot 
# 3.	Boxplot
# 4.	Barplot
# 5.	Pairplot
# 6.	Matriz de correlación 
# 7.	Visualización de la matriz de correlación y covarianza


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Cargamos los datasets
exercise = pd.read_csv('exercise.csv')
car_crashes = pd.read_csv('car_crashes.csv')

exercise.head()
car_crashes.head()
```

```python
exercise.head()
```

```python
# Scatterplot (diagrama de dispersión)
# Para el dataset "exercise"
sns.relplot(
    kind='scatter',
    data=exercise,
    x='time',
    y='pulse',
    hue='kind'
)
plt.show()
```

```python
# Eliminar 'min' de la columna 'time' y convertir a numérico
exercise['time'] = exercise['time'].str.replace('min', '').astype(float)
```

```python
# scatterplot para el dataset "car_crashes"
sns.relplot(
    kind='scatter',
    data=car_crashes,
    x='alcohol',
    y='total',
    hue='abbrev'
)
plt.show()
```

```python
# Joinplot
# Para el dataset "exercise"
sns.jointplot(
    data=exercise,
    x='pulse',
    y='time',
    hue='kind',
)
```

```python
# Para el dataset "car_crashes"
sns.jointplot(
    data=car_crashes,
    x='alcohol',
    y='total',
    hue='abbrev',
)
plt.tight_layout()
plt.show()
```

```python
# Boxplot
# Para el dataset "exercise"
sns.boxplot(
    data=exercise,
    x='kind',
    y='pulse'
)
plt.show()
```

```python
# Para el dataset "car_crashes"
sns.boxplot(
    data=car_crashes,
    x='abbrev',
    y='total'
)
plt.show()
```

```python
# Barplot
# Para el dataset "exercise"
sns.barplot(
    data=exercise,
    x='kind',
    y='pulse'
)
plt.show()
```

```python
# Para el dataset "car_crashes"
sns.barplot(
    data=car_crashes,
    x='abbrev',
    y='total'
)
plt.show()
```

```python
# Pairplot
# Para el dataset "exercise"
sns.pairplot(
    data=exercise,
    hue='kind',
    corner=True,
    kind='scatter'
)
plt.show()
```

```python
# Para el dataset "car_crashes"
sns.pairplot(
    data=car_crashes,
    hue='abbrev',
    corner=True,
    kind='scatter'
)
plt.show()
```

```python
# Matriz de correlación
# Para el dataset "exercise"
from sklearn.preprocessing import StandardScaler

exercise = sns.load_dataset('exercise')
exercise
```

```python
# visualización de la matriz de correlación
sns.pairplot(exercise, hue='kind',corner=True,kind='scatter')
plt.show()
```

```python
# Eliminar 'min' de la columna 'time' y convertir a numérico
exercise['time'] = exercise['time'].str.replace('min', '').astype(float)
```

```python
exercise.columns
```

```python
# Normalizamos los datos
scaler = StandardScaler()
exercise_scaled = scaler.fit_transform(exercise[['time', 'pulse']])
exercise_scaled
```

```python
# Hallar la matriz transpuesta
exercise_scaled.T
```

```python
# Calcular la covarianza de las variables
cov_matrix = np.cov(exercise_scaled.T)
cov_matrix
```

```python
# Graficar la matriz de correlación
plt.figure(figsize=(10, 10))
sns.heatmap(
    cov_matrix,
    cbar=True,
    annot=True,
    square=True,
    fmt='.2f',
    annot_kws={'size': 10},
    xticklabels=['time', 'pulse'],
    yticklabels=['time', 'pulse']
)
plt.tight_layout()
plt.show()
```

```python
# Para el dataset "car_crashes"
car_crashes = sns.load_dataset('car_crashes')
car_crashes
```

```python
# visualización de la matriz de correlación
sns.pairplot(car_crashes, hue='abbrev',corner=True,kind='scatter')
plt.show()
```

```python
# Normalizamos los datos
scaler = StandardScaler()
car_crashes_scaled = scaler.fit_transform(car_crashes[['total', 'speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']])
car_crashes_scaled
```

```python
# Hallar la matriz transpuesta
car_crashes_scaled.T
```

```python
# Calcular la covarianza de las variables
cov_matrix = np.cov(car_crashes_scaled.T)
cov_matrix
```

```python
# Graficar la matriz de correlación
plt.figure(figsize=(10, 10))
sns.heatmap(
    cov_matrix,
    cbar=True,
    annot=True,
    square=True,
    fmt='.2f',
    annot_kws={'size': 10},
    xticklabels=['total', 'speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses'],
    yticklabels=['total', 'speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']
)
```
