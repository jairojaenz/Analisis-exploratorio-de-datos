```python
# Concatenar 3 dataframes
import pandas as pd

# Dataframe 1
df1 = pd.DataFrame({ 'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

# Dataframe 2
df2 = pd.DataFrame({ 'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                        index=[4, 5, 6, 7])

# Dataframe 3
df3 = pd.DataFrame({ 'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])

# Concatenar os 3 dataframes
frames = [df1, df2, df3]
result = pd.concat(frames)
print(result)
```

```python
# asociar claves especificas a cada dataframe
result = pd.concat(frames, keys=['x', 'y', 'z'])
print(result)
```

```python
# seleccioar cada dataframe por clave
print(result.loc['y'])
```

```python
# join outer y join inner
df4 = pd.DataFrame({ 'B': ['B2', 'B3', 'B6', 'B7'],
                        'D': ['D2', 'D3', 'D6', 'D7'],
                        'F': ['F2', 'F3', 'F6', 'F7']},
                        index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=1)
result
```

```python
# join outer y join inner
df4 = pd.DataFrame({ 'B': ['B2', 'B3', 'B6', 'B7'],
                        'D': ['D2', 'D3', 'D6', 'D7'],
                        'F': ['F2', 'F3', 'F6', 'F7']},
                        index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=0)
result
```

```python
# Join inner
result = pd.concat([df1, df4], join='inner')
result
```

```python
# Join inner
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
#usar el indice exacto del dataframe original con reindex
result = pd.concat([df1, df4], axis=1).reindex(df1.index)
result
```

```python
# indexear antes de concatenar
result2 = pd.concat([df1, df4.reindex(df1.index)], axis=1)
result2
```

```python
# ignorar el indice original
result = pd.concat([df1, df4], ignore_index=True, sort=False)
result
```

```python
# ignorar el indice original
result = pd.concat([df1, df4], ignore_index=False, sort=False)
result
```

```python
# si se pasan series sin nombre, se enumeran consecutivamente
s1 = pd.Series(['_0', '_1', '_2', '_3'])
result = pd.concat([df1, s1, s1, s1], axis=1)
result
```

```python
# trabajando con assing
# definir dataframe df1 y serie s1
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], name='X')
# concatenar usando assign
result = df1.assign(X=s1)
result
```

```python
# concatenando con ndims mixtos
# definir dataframe df1 y serie s1
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], name='X')
result = pd.concat([df1, s1], axis=1)
result
```

```python
# al pasar ignore_index=True, se pierden los nombres de las columnas
result = pd.concat([df1, s1], axis=1, ignore_index=True)
result
```

```python
# agregar filas de una serie a un dataframe
s2 = pd.Series(['X0', 'X1', 'X2'], index=['A', 'B', 'C'])
# convertir fila en dataframe
result = pd.concat([df1, s2.to_frame().T],ignore_index=True)
result

```

```python
# Crear un dataframe left y un dataframe right con keys
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],  
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
# concatenar left y right
result = pd.merge(left, right, on = 'key')
result
```

```python
# Crear un dataframe left y un dataframe right con key1 y key2
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
# concatenar left y right
result2 = pd.merge(left, right, on = ['key1', 'key2'])
result2
```

```python
# merge con how left con una llave
result3 = pd.merge(left, right, how='left', on=['key1'])
result3
```

```python
# merge con how left con dos llaves
result3 = pd.merge(left, right, how='left', on=['key1', 'key2'])
result3
```

```python
# merge con how right
result4 = pd.merge(left, right, how='right', on=['key1', 'key2'])
result4
```

```python
# merge con how outer
result5 = pd.merge(left, right, how='outer', on=['key1', 'key2'])
result5
```

```python
# merge con how across
result6 = pd.merge(left, right, how='cross')
result6
```
