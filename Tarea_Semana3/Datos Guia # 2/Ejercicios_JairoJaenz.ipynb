## Ejercicio 6

```python
# 6.	Encontrar en base al archivo “employees.csv” lo siguiente: 
# a.	Generar un dataframe con los datos del fichero
df_employees = pd.read_csv('Ejercicio #6/employees.csv')
df_employees
```

```python
# b. Los empleadores con un salario superior a 30000 y menor a 50000.
df_employees[(df_employees['salary'] > 30000) & (df_employees['salary'] < 50000)]
```

```python
# c. Obtener los registros donde el nombre de un empleado empiece con “J”
import re #Se debe importar re
mask = df_employees.first_name.str.match(r'^[J]', flags=re.IGNORECASE)
df_employees[mask]
```

```python
# d.	Obtener aquellos empleados donde su apellido no contenga la letra “H”.
mask2 = ~df_employees.last_name.str.contains('H')
df_employees[mask2]
```

```python
# e. Ordenar en modo descendente los empleados a partir de su ID de departamento.
df_employees.sort_values(by='department_id', ascending=False)
```

```python
# f.	Obtener el empleado que es gerente de la empresa.
df_employees[df_employees['job_id'] == 'AD_PRES']
```

```python
# g.	Encontrar todos los empleados que están a cargo de un jefe.
df_employees[df_employees['job_id'] == 'AD_VP']
```

```python
# h.	Encontrar los empleados cuyo nombre termina en la letra a.
df_employees[df_employees['first_name'].str.endswith('a')]
```

```python
# i.    Obtener empleados cuyo nombre empiece ya sea por “A”, “L” o “R”, luego ordenar por ID de departamento.
mask3 = df_employees.first_name.str.match(r'^[ALR]', flags=re.IGNORECASE)
df_employees[mask3].sort_values(by='department_id', ascending=True)
```

```python
# j.	Obtener los empleados que pertenezcan ya sea a los departamentos con ID 80 ó 30.
df_employees[(df_employees['department_id'] == 80) | (df_employees['department_id'] == 30)]
```

```python
# k.    Encontrar aquellos empleados que en su nombre la tercera letra es una r.
df_employees[df_employees['first_name'].str[2].str.lower()== 'r']
```

```python
# l.	Obtener los empleados que no pertenezcan a los departamentos con ID 80 ó 30.
df_employees[(df_employees['department_id']!= 80) & (df_employees['department_id']!= 30)]
```

## Ejercicio 9

```python
# 9.	Encontrar basado en el archivo “title.csv” y “cast.csv”
df_titles = pd.read_csv('Ejercicio #9/titles.csv')
df_titles.head()
```

```python
df_cast = pd.read_csv('Ejercicio #9/cast.csv')
df_cast.head()
```

```python
# a.	¿Cuántas películas están listadas por el nombre "Drácula”?
df_titles[df_titles['title'] == 'Dracula'].shape[0]
```

```python
# b.	Títulos más comunes en la historia filmográfica
df_titles['title'].value_counts().head(10)
```

```python
# c.	¿Cuál fue la primer película hecha titulada "Romeo and Juliet”?
df_titles[df_titles['title'] == 'Romeo and Juliet'].head(1)
```

```python
df_titles = df_titles.fillna(1900)
df_titles
```

```python
# d.	 Listar todas las películas que contengan la palabra "Exorcist" y ordenadas de la más antigua a la más reciente.
mask4=df_titles.title.str.contains('Exorcist')
df_titles[mask4.astype(bool)].sort_values(by='year', ascending=True)
```

```python
#e.	¿Cuántas películas fueron realizadas en el año 1950?
peliculas_1950 = df_titles[df_titles['year'] == 1950].shape[0]
print(f'Peliculas que fueron realizadas en el año 1950 :  {peliculas_1950}')
```

```python
# a.	¿Cuántas películas fueron hechas de 1950 a 1959?
peliculas_1950_a_1959 = df_titles[(df_titles['year'] >= 1950) & (df_titles['year'] <= 1959)].shape[0]
print(f'Peliculas que fueron hechas de 1950 a 1959 :  {peliculas_1950_a_1959}')
```

```python
# g.	¿En qué año se presentó la película llamada "Batman”?
df_titles[df_titles['title'] == 'Batman'].head(1)
```

```python
# h.	¿Cuántos roles o papeles hubo en la película "The Godfather"?
print(f'Hubieron {df_cast[df_cast['title'] == 'The Godfather'].shape[0]} roles o papeles hubo en la película "The Godfather')
```

```python
# j.	¿Cuántos papeles en la película "The Godfather" si estuvieron clasificados con un valor "n"?
print(f'{df_cast[(df_cast['title'] == 'The Godfather') & (df_cast['n']!=None)].shape[0]} papeles estuvieron clasificados con valor de "n" en la pelicula "The GodFather')
```

```python
# k.	Mostrar el elenco completo de la película "2001: A Space Odyssey" ordenado por su clasificación 
# "n", ignorando los papeles que no se les asigno ningún valor "n"
df_cast[df_cast['title'] == '2001: A Space Odyssey'].sort_values(by='n').dropna()
```

```python
# l.	Mostrar el elenco completo ordenado por la clasificación "n" de la película "Drácula" de 1958.
df_cast[df_cast['title'] == 'Dracula'][df_cast['year'] == 1958].sort_values(by='n').dropna()
```

```python
# m.	¿Cuántos papeles fueron listados en la película "The Wizard of Oz" de 1939?
df_cast[df_cast['title'] == 'The Wizard of Oz'][df_cast['year'] == 1939].shape[0]
```

```python
# n.	¿Cuántas gente ha hecho el papel de "Romeo"?
df_cast[df_cast['character']== 'Romeo'].shape[0]
```

```python
# o.	¿Cuántos papeles ha hecho "Robert De Niro" en su carrera?
df_cast[df_cast['character'] == 'Robert De Niro'].shape[0]
```

```python
# p.	Listado de papeles de soporte que tuvo el actor "Charlton Heston" en la década de los 50's, ordenado por año. 
df_cast[df_cast['name'] == 'Charlton Heston'][df_cast['year'].between(1950,1959)].sort_values(by='year').dropna()
```

```python
# q.	Listado de papeles como protagonista que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente. 
df_cast[df_cast['name'] == 'Charlton Heston'][df_cast['year'].between(1960,1969)].sort_values(by='year', ascending=False).dropna()
```

```python
# r.	¿Cuántos papeles para actores hubo en la década de los 50's?
df_cast[df_cast['type']=='actor'][df_cast['year'].between(1950,1959)].shape[0]
```

```python
# s.	¿Cuántos papeles para actrices hubo en la década de los 50's?
df_cast[df_cast['type']=='actress'][df_cast['year'].between(1950,1959)].shape[0]
```

```python
# t.	¿Cuántos papeles protagónicos existen desde el principio de la historia filmográfica hasta 1970?
df_cast[df_cast['year']<= 1970][df_cast['n']==1].shape[0]
```
