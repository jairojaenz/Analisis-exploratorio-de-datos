```python
import pandas as pd
```

```python
url = 'https://es.wikipedia.org/wiki/Comunidad_aut%C3%B3noma'
df_superficie = pd.read_html(url)[4]
df_superficie.columns = df_superficie.columns.get_level_values(1)
df_poblacion = pd.read_html(url)[5]
df_poblacion.columns = df_poblacion.columns.get_level_values(1)
```

```python
# crear copia de ambos dataframes
df_superficie_copy = df_superficie.copy()
df_poblacion_copy = df_poblacion.copy()
```

```python
# Elimine la última fila de totales en cada DataFrame
df_superficie_copy = df_superficie_copy.drop(df_superficie_copy.index[-1])
df_superficie_copy
```

```python
df_poblacion_copy = df_poblacion.drop(df_poblacion_copy.index[-1])
df_poblacion_copy
```

```python
# Elimimar la primera columna del dataframe de superficie por su indice
df_superficie_copy = df_superficie_copy.drop(df_superficie_copy.columns[0], axis=1)
# Cambie el nombre de la columna Comunidad Autónoma por Comunidad
df_superficie_copy = df_superficie_copy.rename(columns={'Comunidad o ciudad autónoma.1': 'Comunidad'})
df_superficie_copy = df_superficie_copy.rename(columns={'Superficie (km²)': 'Superficie'})
# Elimine la columna porcentaje de la tabla de superficie
df_superficie_copy = df_superficie_copy.drop(columns=['Porcentaje'])
df_superficie_copy
```

```python
# Elimimar la primera columna del dataframe de población por su indice
df_poblacion_copy = df_poblacion_copy.drop(df_poblacion_copy.columns[0], axis=1)
# cambie el nombre de la columna Comunidad o ciudad autónoma por Comunidad
df_poblacion_copy = df_poblacion_copy.rename(columns={'Comunidad o ciudad autónoma.1': 'Comunidad'})
# cambiar el nombre de la columna Densidad (hab./km²) por Densidad
df_poblacion_copy = df_poblacion_copy.rename(columns={'Densidad (hab./km²)': 'Densidad'}) 
# Elimine la columna porcentaje de la tabla de población
df_poblacion_copy = df_poblacion_copy.drop(columns=['Porcentaje'])
df_poblacion_copy
```

```python
# Reemplace los valores de población y superficie para que sean números y convierta las columnas a entero.
df_poblacion_copy['Población'] = df_poblacion_copy['Población'].str.replace('\xa0', '').astype(int)
df_poblacion_copy
```

```python
df_poblacion_copy.dtypes
```

```python
# Convertir la columna Superficie a entero (es un object)
df_superficie_copy['Superficie'] = df_superficie_copy['Superficie'].str.replace('\xa0', '').astype(int)
df_superficie_copy
```

```python
df_superficie_copy.dtypes
```

```python
# Realice la mezcla de población y superficie en un único DataFrame filas y columnas
df_copy = pd.merge(df_superficie_copy, df_poblacion_copy, on='Comunidad', how='inner')
df_copy
```

```python
# Calcule la densidad de población de cada comunidad autónoma.
df_copy['Densidad'] = df_copy['Población'] / df_copy['Superficie']
df_copy
```

```python
# Los ficheros emisiones-2016.csv, emisiones-2017.csv, emisiones-2018.csv y emisiones-2019.csv,
# contienen datos sobre las emisiones contaminantes en la ciudad de Madrid en los años 2016, 2017,
# 2018 y 2019 respectivamente. Escribir un programa con los siguientes requisitos:
# 1. Generar un DataFrame con los datos de los cuatro ficheros.
# 2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, ..., D31.
# 3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.
# 4. Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).
# 5. Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones, contaminantes y fecha.
# 6. Mostrar por pantalla las estaciones y los contaminantes disponibles en el DataFrame.

import numpy as np
import pandas as pd
import datetime

# 1. Generar un DataFrame con los datos de los cuatro ficheros.
df_2016 = pd.read_csv('Ejercicio_1/emisiones-2016.csv', sep=';')
df_2017 = pd.read_csv('Ejercicio_1/emisiones-2017.csv', sep=';')
df_2018 = pd.read_csv('Ejercicio_1/emisiones-2018.csv', sep=';')
df_2019 = pd.read_csv('Ejercicio_1/emisiones-2019.csv', sep=';')

df = pd.concat([df_2016, df_2017, df_2018, df_2019])
df
```

```python
# 2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, ..., D31.
df_copy = df.copy()
df_copy = df_copy[['ESTACION', 'MAGNITUD', 'ANO', 'MES', 'D01', 'D02', 'D03', 'D04', 'D05', 'D06', 'D07', 'D08', 'D09', 'D10', 'D11', 'D12', 'D13', 'D14', 'D15', 'D16', 'D17', 'D18', 'D19', 'D20', 'D21', 'D22', 'D23', 'D24', 'D25', 'D26', 'D27', 'D28', 'D29', 'D30', 'D31']]
df_copy
```

```python
# 3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.
df_copy = pd.melt(df_copy, id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')
df_copy
```

```python
# 4. Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).
df_copy['FECHA'] = df_copy['ANO'].astype(str) + '-' + df_copy['MES'].astype(str) + '-' + df_copy['DIA'].str[1:]
df_copy['FECHA'] = pd.to_datetime(df_copy['FECHA'], format='%Y-%m-%d', errors='coerce')
df_copy
```

```python
# 5. Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones, contaminantes y fecha.
df_copy = df_copy.drop(df_copy[np.isnat(df_copy['FECHA'])].index)
df_copy = df_copy[['ESTACION', 'MAGNITUD', 'FECHA']]
df_copy
```

```python
# 6. Mostrar por pantalla las estaciones y los contaminantes disponibles en el DataFrame.
print('Estaciones:', df_copy['ESTACION'].unique())
print('Contaminantes:', df_copy['MAGNITUD'].unique())
```

```python
# 7 Crear una función que reciba una estación, un contaminante y un rango de fechas y
#devuelva una serie con las emisiones del contaminante dado en la estación y rango de
# fechas dado.
def evolucion_emisiones(estacion, contaminante, fecha_inicio, fecha_fin):
    return df_copy[(df_copy['ESTACION'] == estacion) & (df_copy['MAGNITUD'] == contaminante) & (df_copy['FECHA'] >= fecha_inicio) & (df_copy['FECHA'] <= fecha_fin)].sort_values('FECHA')

evolucion_emisiones(56, 8, datetime.datetime.strptime('2018-01-01', '%Y-%m-%d'), datetime.datetime.strptime('2018-01-31', '%Y-%m-%d'))
```

```python
# 10 Crear una función que reciba una estación y un contaminante y devuelva un resumen
# descriptivo de las emisiones del contaminante indicado en la estación indicada.
df = df[['ESTACION', 'MAGNITUD', 'ANO', 'MES', 'D01', 'D02', 'D03', 'D04', 'D05', 'D06', 'D07', 'D08', 'D09', 'D10', 'D11', 'D12', 'D13', 'D14', 'D15', 'D16', 'D17', 'D18', 'D19', 'D20', 'D21', 'D22', 'D23', 'D24', 'D25', 'D26', 'D27', 'D28', 'D29', 'D30', 'D31']]
def resumen_estacion(estacion, contaminante):
    return df[(df['ESTACION'] == estacion) & (df['MAGNITUD'] == contaminante)].describe()

resumen_estacion(56, 8)
```

```python
# 11 Crear una función que devuelva las emisiones medias mensuales de un contaminante y un
# año dados para todas las estaciones.
def evolucion_mensual(contaminante, año):
    return df[(df['MAGNITUD'] == contaminante) & (df['ANO'] == año)].groupby('MES').mean()

evolucion_mensual(8, 2019)
```

```python
# 12 Crear un función que reciba una estación de medición y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.
def medias_mensuales_por_estacion(estacion):
    return df[df['ESTACION'] == estacion].groupby(['MES', 'MAGNITUD']).mean().reset_index()

medias_mensuales_por_estacion(56)
```
